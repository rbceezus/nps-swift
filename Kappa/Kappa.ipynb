```python
import pandas as pd
from sklearn.metrics import cohen_kappa_score

# 1. Carregar o arquivo CSV
# Certifique-se de que o ficheiro está na mesma pasta do seu script Python
nome_arquivo = "anotacoes_amostra.xlsx"
df = pd.read_excel(nome_arquivo)

# 2. Limpeza rápida (opcional, mas recomendada): 
# Remove linhas onde algum dos anotadores deixou em branco
df = df.dropna(subset=['categoria_anotador_1', 'categoria_anotador_2'])

# 3. Extrair os dados de cada avaliador
avaliador_a = df['categoria_anotador_1']
avaliador_b = df['categoria_anotador_2']

# 4. Calcular o coeficiente Kappa de Cohen
kappa = cohen_kappa_score(avaliador_a, avaliador_b)

# 5. Exibir os resultados
print(f"Total de comentários avaliados: {len(df)}")
print(f"Coeficiente Kappa de Cohen: {kappa:.4f}")

# 6. Interpretação automática do resultado
print("Nível de Concordância: ", end="")
if kappa < 0:
    print("Nenhuma (pior que o acaso)")
elif kappa <= 0.20:
    print("Ínfima / Insignificante")
elif kappa <= 0.40:
    print("Razoável / Discreta")
elif kappa <= 0.60:
    print("Moderada")
elif kappa <= 0.80:
    print("Substancial")
else:
    print("Quase Perfeita")
```
