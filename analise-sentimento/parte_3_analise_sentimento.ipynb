# Projeto NPS Swift - Entrega 1: Parte 3 - Definição da Análise de Sentimento

Este notebook foca exclusivamente na definição estratégica da análise de sentimento, conforme os requisitos estabelecidos para o projeto.

## 3. Definição da Análise de Sentimento

Nesta etapa, definimos como o problema de sentimento será tratado, considerando a granularidade, a fonte de verdade (ground truth) e o tratamento de casos complexos.

### 3.1 — Granularidade dos sentimentos

| Abordagem | Classes | Decisão |
| :--- | :--- | :--- |
| **Ternária** | Positivo / Neutro / Negativo | **Escolhida** |

**Justificativa:** 
Com base na EDA realizada anteriormente, observamos que uma parcela significativa dos clientes (aproximadamente 16%) apresenta um perfil neutro ou indiferente. A abordagem ternária é a mais adequada para o NPS, pois permite distinguir entre o cliente que está ativamente satisfeito/insatisfeito e aquele que não possui uma opinião forte, o que é crucial para as ações de retenção da Swift.

### 3.2 — O problema da nota vs. o texto

**Decisão:** Usar o **Texto como fonte de verdade (Ground Truth)**.

**Justificativa:** 
Embora a nota numérica seja um proxy rápido, identificamos na EDA diversos casos de discordância (ex: nota 10 com comentário reclamando de preço, ou nota 3 elogiando o atendimento mas criticando a logística). Para o treinamento do modelo de NLP, o texto reflete com maior precisão a experiência real relatada pelo cliente. Validaremos manualmente os casos onde a nota e o texto divergem para garantir um dataset de treino robusto.

### 3.3 — Tratamento de comentários mistos

**Estratégia:** Classificação pelo **Sentimento Dominante** ou **Pior Experiência**.

**Definição:** 
Comentários que apresentam sentimentos conflitantes (ex: "Atendimento nota 10, mas o preço está absurdo") serão tratados da seguinte forma:
1. Se houver um relato de falha operacional grave (detrator), o sentimento será classificado como **Negativo**, priorizando a identificação de pontos de melhoria.
2. Caso contrário, será avaliado o sentimento que ocupa a maior parte do texto.

**Análise da Amostra:** Na amostra anotada de 200 comentários, identificamos que aproximadamente 5% dos registros apresentam esse padrão misto.
